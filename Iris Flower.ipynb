```python
x=df.drop(columns='Species',axis=1)
y=df['Species']
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
X_train
```

```python
X_test
```

```python
print(x.shape,X_train.shape,X_test.shape)
```

```python
print('training data shape is :{}'.format(X_train.shape))
print('training label shape is :{}'.format(y_train.shape))
print('testing data shape is :{}'.format(X_test.shape))
print('testing label shape is :{}'.format(y_test.shape))
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.svm import SVC
model_svm=SVC()
model_svm.fit(X_train,y_train)
```

```python
pred=model_svm.predict(X_test)
pred
```

```python
accuracy_score(y_test,pred)
```

```python
print("the matrix of confusion:")
print(confusion_matrix(y_test,pred))
```

```python
print("the classification report:")
print(classification_report(y_test,pred))
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
predictions=logmodel.predict(X_test)
percentage=logmodel.score(X_test,y_test)
r=confusion_matrix(y_test,predictions)
print("Confusion Matrix")
print(r)
```

```python
print(logmodel.score(X_train,y_train))
```
